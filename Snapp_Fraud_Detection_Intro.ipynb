```python
# Snapp location analysis for fraud detection in September

df = pd.read_csv('df_location.csv')
```

```python
df.rename(columns={'Date/Time':'Date'}, inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Date', 'Base'], ascending=[True, True])
df = df.reset_index(drop=True)
df['Hour'] = df['Date'].dt.hour
df['Day'] = df['Date'].dt.day
df['WeekDay'] = df['Date'].dt.day_name()
df['Coordinate'] = df['Lat'].round(4).astype(str).apply(lambda x: '{0:0<8}'.format(x)) + ' ' + df['Lon'].round(4).astype(str).apply(lambda x: '{0:0<8}'.format(x))
df.drop(columns=['Date','Lat','Lon'],inplace=True)
df = df[['Day', 'WeekDay','Hour','Coordinate','Base']]
```

```python
# Hour

for hour in df['Hour'].unique():
    h1 = df.loc[df['Hour'] == hour]
    print('********','Hour:',hour,'********')
    print('Request Chance:',round(h1['Hour'].value_counts().sum()/df.shape[0]*100,2),'%\n')
    print('Affiliation Rate % :')
    print(round(h1['Base'].value_counts(normalize=True)*100,1),'\n')
    print('The Most Popular Pick up Location:')
    print(h1['Coordinate'].value_counts().index[0])
    print('_________________________________________________________________','\n')
```

```python
# WeekDay

for day in df['WeekDay'].unique():
    h1 = df.loc[df['WeekDay'] == day]
    print('********','Week Day:',day,'********')
    print('Requests Percentage of total:',round(h1['WeekDay'].value_counts().sum()/df.shape[0]*100,1),'%\n')
    print('Affiliation Rate % :')
    print(round(h1['Base'].value_counts(normalize=True)*100,1),'\n')
    print('The Most Popular Pick up Location:')
    print(h1['Coordinate'].value_counts().index[0])
    print('_________________________________________________________________','\n')
```
