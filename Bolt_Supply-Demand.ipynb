```python
df = pd.read_csv('df.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Show null values and duplicates
total = df.isnull().sum().sort_values(ascending=False)
percent = round((df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)*100, 3)
missing_data = pd.concat([total, percent], axis=1, keys=['NullTotal', 'Null%'])
for dup in missing_data.index:
    missing_data.at[dup,'DupTotal'] = df.duplicated(subset=[dup]).sum()
    missing_data.at[dup,'Dup%'] = round((df.duplicated(subset=[dup]).sum())/df.shape[0]*100,3)
missing_data
```

```python
# 36 lowest Supply/Demand  in each week, count values top 5. Is it a good determinant od undersupply or not?

for week in df['Week'].unique():
    print('Week',week,':')
    print(df.sort_values(['SupplyDemand'], ascending=[True]).loc[(df['Week'] == week)][:36]['Hour'].value_counts()[:5])
    print('_______________________________________\n')
```

```python
# Rush hours (8,9,18,19) considering all 5 weeks

df2 = df.loc[(df['Hour']==8)|(df['Hour']==9)|(df['Hour'] ==18)|(df['Hour']==19)]
dfresult = pd.DataFrame(data={'Drivers':[''],},columns=['Drivers'])

dfresult['Drivers'] = round(df2['Drivers'].mean(),0)
dfresult['Online(H)'] =round(df2['Online(H)'].mean(),0)
dfresult['OnlinePerDrivers'] =round(df2['OnlinePerDrivers'].mean(),1)
dfresult['Booking(H)'] =round(df2['Booking(H)'].mean(),0)
dfresult['Waiting(H)'] =round(df2['Waiting(H)'].mean(),0)
dfresult['Rides'] =round(df2['Rides'].mean(),0)
dfresult['RidesPerOnline'] =round(df2['RidesPerOnline'].mean(),2)
dfresult['PeopleSaw0Cars'] =round(df2['PeopleSaw0Cars'].mean(),0)
dfresult['PeopleSaw+1Cars'] =round(df2['PeopleSaw+1Cars'].mean(),0)
dfresult['Demand'] =round(df2['Demand'].mean(),0)
dfresult['CoverageRatio%'] =round(df2['CoverageRatio%'].mean(),0)
dfresult['Supply'] =round(dfresult['Waiting(H)']/dfresult['OnlinePerDrivers'].mean(),0)
dfresult['SupplyDemand'] =round(dfresult['Supply']/dfresult['Demand'].mean(),1)

dfresult
```

**Approach 2 (Demand-friendly):** RPH(0.81) multiply 8 eur = 6.48 eur but drivers will lose money in non-rush hours, aound 50% increase in drivers means increase in Online(H) in hours that we have oversupply. Then RidesPerOnine will decrease, so the drivers income.
