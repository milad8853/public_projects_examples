```python
df1 = pd.read_csv('campaigns.csv')
df2 = pd.read_csv('events_table.csv')
df2.drop(columns=['id'], inplace=True)
df = df1.merge(df2, on="campaign_id")
df['ts'] = pd.to_datetime(df['ts'])
```

```python
# Count Values
def count(df):
    pd.set_option('display.max_rows', 1000)
    for column in df.columns:
        print(column,':')
        display(df[column].value_counts(dropna=False))
        print('___________________________________','\n')
count(df)
```

```python
# Show null values and duplicates
def nulldup(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = round((df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)*100, 3)
    missing_data = pd.concat([total, percent], axis=1, keys=['NullTotal', 'Null%'])
    for dup in missing_data.index:
        missing_data.at[dup,'DupTotal'] = df.duplicated(subset=[dup]).sum()
        missing_data.at[dup,'Dup%'] = round((df.duplicated(subset=[dup]).sum())/df.shape[0]*100,3)
    display(missing_data)
nulldup(df)	
```

```python
dfresult = pd.DataFrame(data={'campaign_id':[''],'country':[''],'company_id':[''],'events':['']},
                        columns=['campaign_id','country','company_id','events'])
index=0
for campaign in df['campaign_id'].unique():
    dftarget = df.loc[df['campaign_id'] == campaign]
    dfresult.at[index,'campaign_id'] = campaign
    dfresult.at[index,'country'] = dftarget['country'].mode()
    dfresult.at[index,'company_id'] = dftarget['company_id'].unique().sum()
    dfresult.at[index,'events'] = round(dftarget['event'].sum()/dftarget.shape[0]*100,2)
    index=index+1
dfresult = dfresult.sort_values('events', ascending=False)
dfresult
```

```python
dfresult2 = pd.DataFrame(data={'company_id':[''],'events':['']}, columns=['company_id','events'])
index=0
for cid in dfresult['company_id'].unique():
    dftarget2 = dfresult.loc[dfresult['company_id'] == cid]
    dfresult2.at[index,'company_id'] = cid
    dfresult2.at[index,'events'] = round(dftarget2['events'].mean(),2)
    index=index+1
dfresult2.sort_values('company_id', ascending=False)
```
